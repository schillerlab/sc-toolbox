### Remove certain cell types

```python
import numpy as np    ## have to use this to remove nan values

column = "cell type"
to_remove = [np.nan, "Mki67+ Proliferating cells (mainly T cells)", "MKi67+ Proliferating cells"]
marker = marker[~marker.loc[:, column].isin(to_remove)].copy()
```

```python
codes.plot_marker_dendrogram(marker, thresh = 0.7, column = "cell type", label_size = 12,
                             figsize = (20, 6), save = None)
```

### Plot Dendrogram only for certain subset

```python
cell_types = ["Ciliated cells", "Goblet cells", "Club cells", "Club cells (d21)",  "AT2 cells", 
              "AT2 cells early after injury (d3/d7)", "AT1 cells", "Ciliated cell subset"]
marker_subset = marker[marker.loc[:, column].isin(cell_types)].copy()

codes.plot_marker_dendrogram(marker_subset, thresh = 0.7, column = "cell type", label_size = 20,
                             orient = "left", figsize = (8, 6), save = None)
```
